#Data Visualization

```python
df['accent'].value_counts()
```

```python
# Prepare data
accent_counts = df['accent'].value_counts().reset_index()
accent_counts.columns = ['accent', 'count']
accent_counts = accent_counts.sort_values('accent')  # optional sorting

# Plot
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(data=accent_counts, x='accent', y='count', palette='Paired')
plt.title('Distribution of Accents Before Sampling', fontsize=14)
plt.xlabel('Accent', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['age'].value_counts()
```

```python
# Value counts and sort by age if needed
age_counts = df['age'].value_counts().sort_index()

# Define a modern color palette
colors = plt.cm.Pastel2.colors

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(age_counts.index.astype(str), age_counts.values, color=colors[:len(age_counts)], edgecolor='black')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 200, str(height), ha='center', fontsize=10)

# Labels and title
plt.title('Distribution of Age Before Sampling', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Drop MFCC columns
non_mfcc_df = df[[col for col in df.columns if 'mfcc' not in col.lower()]]

# Optional: drop non-numeric columns like 'gender' or 'unique_id' before correlation
numeric_df = non_mfcc_df.select_dtypes(include=[np.number])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5, cbar_kws={'shrink': 0.75})
plt.title('Feature Correlation Heatmap (Excluding MFCCs)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Filter only MFCC-related columns
mfcc_df = df[[col for col in df.columns if 'mfcc' in col.lower()]]

# Compute correlation matrix
mfcc_corr = mfcc_df.corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(mfcc_corr, annot=False, cmap='viridis', fmt=".2f", square=True, linewidths=0.5, cbar_kws={'shrink': 0.75})
plt.title('Correlation Heatmap of MFCC_Mean Features', fontsize=14)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Filter only MFCC std columns
mfcc_std_df = df[[col for col in df.columns if 'mfcc' in col.lower() and 'std' in col.lower()]]

# Compute correlation matrix
mfcc_std_corr = mfcc_std_df.corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(mfcc_std_corr, annot=False, cmap='viridis', fmt=".2f", square=True, linewidths=0.5, cbar_kws={'shrink': 0.75})
plt.title('Correlation Heatmap of MFCC STD Features', fontsize=14)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
